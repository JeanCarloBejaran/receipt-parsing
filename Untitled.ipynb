# — DATE

```python
def _find_date(text):
    if not text:
        return None
    if 'FECHA' in text:
        date = text.split('FECHA')[1]
        date = date.strip(":")
        date = re.sub(r'\s', '', date)
        date = dateutil.parser.parse(date)
        return date
    else:
        return None
    
def get_date(blocktype):
    date = df.loc[(df.BlockType == blocktype), "Text"].apply(_find_date).dropna().values[0]
    if date:
        date = pd.to_datetime(date).strftime('%Y%m%d')
        print("Date found: {}".format(date))
        return date
    else:
        print("Date NOT found!!!")
```

# RNC

```python
def _find_rnc(text):
    if not text:
        return None
    if 'RNC' in text:
        rnc = text.split('RNC')[1]
        rnc = rnc.strip(":")
#         rnc = re.sub(r'\s', '', rnc)
        rnc = re.findall('(?:\d-?){9}', rnc)
        if rnc:
            rnc = re.sub(r'-', '', rnc[0])
            return rnc
#     else:
# #         rnc = re.sub(r'\s', '', text)
#         rnc = re.findall('(?:\d-?){9}$', text)
#         if rnc:
#             rnc = re.sub(r'-', '', rnc[0])
#             return rnc
    else:
        return None

def get_rnc(blocktype):
    rnc_list = df.loc[(df.BlockType == blocktype), "Text"].apply(_find_rnc).dropna().values.tolist()
    if rnc_list:
        print("RNC's found: {}".format(rnc_list))
        return rnc_list
    else:
        print("RNC's NOT found!!!")
```
